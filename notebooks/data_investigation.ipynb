```python
bins = torch.zeros((lt.n_words))
```

```python
for d in tqdm(dataset['train']):
    bins += np.bincount(lt.encode(d['text'],False, False), minlength=lt.n_words)
```

```python
sibins = torch.argsort(bins, descending=True)
sbins = bins[sibins]
```

```python
tokens = [lt.sp_model.IdToPiece(i) for i in sibins.tolist()]
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

def plot_log_linear_bar(x, y, title="Log-Linear Bar Chart", xlabel="X", ylabel="Y"):
    p = figure(title=title, x_axis_label=xlabel, x_range=x,
               y_axis_label=ylabel, y_axis_type="log",
              height=350, width=1600)
    p.vbar(x=x,bottom=1, top=y, width=0.5)
    p.xaxis.major_label_orientation = 45

    return p
```
